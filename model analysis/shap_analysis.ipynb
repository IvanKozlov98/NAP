<h1>Model analysis</h1>

### Building milticlass classification model based on 2R chromosome:

1) Make dataset based on DNA.

```python
!cd ../ && python -m src.preprocessing -c 2R -r 50000 -rp data/Amer/Amer_repeat_annotation_total.bed -chr data/Amer/2R/2R_chr.fa -g data/Amer/2R/2r_annotation_order.txt -gm data/Amer/2R/ncounts_2R.tsv -o 2R_dataset.csv
```

2) Extract target for model from hic-data

```python
!cd ../ && python -m src.extract_target -c 2R -r 50000 -hic https://genedev.bionet.nsc.ru/site/hic_out/by_Project/Anopheles/ActualData/hic/AmerR4_V4/AmerR4A_V4.1000.hic -m Multiclass -o 2R_multi_target_50000.npy
```

3) Build model based on data from preprocessing.py and extract_target.py

```python
!cd ../ && python -m src.training -x 2R_dataset.csv -y 2R_multi_target_50000.npy -t 4 -m Multiclass -o 2R_model_50000_multi.bin
```

### Test this model on 3L chromosome:

1) Prepare dataset and target

```python
!cd ../ && python -m src.preprocessing -c 3L -r 50000 -rp data/Amer/Amer_repeat_annotation_total.bed -chr data/Amer/3L/3L_chr.fa -g data/Amer/3L/3l_annotation_order.txt -gm data/Amer/3L/ncounts_3L.tsv -o 3L_dataset.csv
```

2) Prediction hic-data based on dataset from preprocessing.

```python
!cd ../ && python -m src.predictor -d 3L_dataset.csv -model 2R_model_50000_multi.bin -m Multiclass -o 3L_multi_prediction.npy
```
